```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Indlæser data fra rød- og hvidvinfilerne
white_wine_path = "https://raw.githubusercontent.com/foxdocs/cph-bi-2024/main/Data/UCL-Wine/winequality-white.csv"
red_wine_path = "https://raw.githubusercontent.com/foxdocs/cph-bi-2024/main/Data/UCL-Wine/winequality-red.csv"
```

```python
# Indlæser datasættene i pandas DataFrames
red_wine = pd.read_csv(red_wine_path, sep=';')
white_wine = pd.read_csv(white_wine_path, sep=';')
```

```python
# Tilføjer en kolonne for at identificere vintypen
red_wine['wine_type'] = 'red'
white_wine['wine_type'] = 'white'
```

```python
# Tjekker for manglende værdier i begge datasæt
missing_values_red = red_wine.isnull().sum()
missing_values_white = white_wine.isnull().sum()
```

```python
# Kombinerer de to datasæt til ét, hvor vintypen bevares
combined_wine = pd.concat([red_wine, white_wine], ignore_index=True)
```

```python
# Tjekker antallet af rækker, kolonner og datatyper i det kombinerede datasæt
rows, cols = combined_wine.shape
data_types = combined_wine.dtypes
```

```python
plt.figure(figsize=(19, 7))
plt.subplot(1, 2, 2)
sns.barplot(x=density_10.index.astype(str), y=density_10.values)
plt.title('Density of pH Bins (10 Bins)')
plt.tight_layout()
plt.show()
```


### Skab et Heatmap eller Korrelationsmatrix

```python
# Calculate the correlation matrix, excluding non-numeric columns
# Select only numeric columns for correlation
numeric_columns = combined_wine.select_dtypes(include=[float, int])
red_correlation = red_wine.select_dtypes(include=[float, int])
white_correlation = white_wine.select_dtypes(include=[float, int])
```

```python
# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()
```

```python
# Plot the heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Matrix of Wine Attributes')
plt.show()
```

```python
# Display the highest and lowest correlations with quality
correlation_with_quality = correlation_matrix['quality'].sort_values(ascending=False)
print(correlation_with_quality)
```

```python
# Calculate correlation matrices for red and white wines separately
correlation_matrixred = red_correlation.corr()
correlation_matrixwite = white_correlation.corr()
```

```python
red_correlation_quality = correlation_matrixred['quality'].sort_values(ascending=False)
white_correlation_quality = correlation_matrixwite['quality'].sort_values(ascending=False)
```

```python
# Red wine correlation heatmap
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.heatmap(correlation_matrixred, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Matrix - Red Wine')
plt.tight_layout()
plt.show()
```

```python
# with wine correlation heatmap
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
sns.heatmap(correlation_matrixwite, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Matrix - White Wine')
plt.tight_layout()
plt.show()
```

```python
# Print correlations with quality for both red and white wines
print("Rødvin korrelation med kvalitet:")
print(red_correlation_quality)
print("\nHvidvin korrelation med kvalitet:")
print(white_correlation_quality)
```

### Rødvin: Alkohol har den største positive effekt på kvalitet (0.48),
### flygtig syre har en stærk negativ effekt (-0.39).
### Hvidvin: Alkohol har også en stærk positiv effekt (0.44),
### men andre attributter som pH har en større positiv indflydelse sammenlignet med rødvin.

```python
# Identificer outliers i 'residual sugar' ved hjælp af IQR-metoden
Q1 = combined_wine['residual sugar'].quantile(0.25)
Q3 = combined_wine['residual sugar'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Definer outlier-grænser
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Find rækker med outliers
outliers = combined_wine[(combined_wine['residual sugar'] < lower_bound) | 
                         (combined_wine['residual sugar'] > upper_bound)]
```

```python
cleaned_data = combined_wine.drop(outliers.index)
```

```python
print(f"Antal outliers fundet: {len(outliers)}")
print("Rækker med outliers i 'residual sugar':")
print(outliers)
```

```python
# Tjek eksisterende kolonner i datasættet
print("Eksisterende kolonner i datasættet:")
print(combined_wine.columns)
```

```python
# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()
```

```python
# F# Find attributter, der har lav korrelation med kvalitet (< |0.05|)
low_corr_attributes = correlation_matrix['quality'][abs(correlation_matrix['quality']) < 0.05].index.tolist()
```

```python
print("Attributter med lav korrelation til kvalitet:", low_corr_attributes)
```

```python
# Fjern attributterne med lav korrelation til kvalitet
final_data = numeric_columns.drop(columns=low_corr_attributes)
```

```python
correlation_matrix_final = final_data.corr().abs()
```

```python
# Find den øverste trekant af korrelationsmatrixen for at identificere høje korrelationer
upper_triangle = correlation_matrix_final.where(np.triu(np.ones(correlation_matrix_final.shape), k=1).astype(bool))
```

```python
# Identificer kolonner, hvor korrelationen er over 0.8
high_corr_attributes = [column for column in upper_triangle.columns if any(upper_triangle[column] > 0.8)]
```

```python
print("Attributter med høj korrelation med andre attributter:", high_corr_attributes)
```

```python
# Fjern de højt korrelerede kolonner
final_data = final_data.drop(columns=high_corr_attributes)
```

```python
print("Endeligt datasæt efter fjernelse af lavt og højt korrelerede attributter:")
print(final_data.head())
```

```python
# Tjek eksisterende kolonner for at finde kategoriske data
print("Kolonner i datasættet:", final_data.columns)
```

```python
# Anvend one-hot encoding på 'wine_type' kolonnen
# Tjekker om 'wine_type' kolonnen er i datasættet
if 'wine_type' in final_data.columns:
    # Konverter 'wine_type' til numerisk med one-hot encoding
    final_data_encoded = pd.get_dummies(final_data, columns=['wine_type'], drop_first=True)
    print("Kategorisk kolonne 'wine_type' er transformeret til numerisk.")
else:
    print("Kolonnen 'wine_type' findes ikke i datasættet.")
```
